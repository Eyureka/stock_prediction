```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timezone
from dateutil.relativedelta import relativedelta

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
current_date = datetime.today()
current_date = datetime.strftime(current_date, "%Y-%m-%d")
current_date = datetime.strptime(current_date, "%Y-%m-%d").replace(tzinfo=timezone.utc)
five_year_previous_date = current_date - relativedelta(years=5)
current_date_timestamp = str(int(datetime.timestamp(current_date)))
five_year_previous_timestamp = str(int(datetime.timestamp(five_year_previous_date)))
```

## fetching the dataset from yahoo finance site using api

```python
company_code = "INFY.NS"
dataset_url = "https://query1.finance.yahoo.com/v7/finance/download/"+company_code+"?period1="+five_year_previous_timestamp+"&period2="+current_date_timestamp+"&interval=1d&events=history&includeAdjustedClose=true"
date_parse = lambda date_str: datetime.strptime(date_str, "%Y-%m-%d")
stock_df = pd.read_csv(dataset_url, index_col="Date", date_parser=date_parse)
stock_df
```

#### feature engineering- removing null values, extracting date and open price columns.
#### open price is chosen since it depends on the previous closing price, thus can provide better predictions

```python
stock_df[stock_df["Open"].isnull()]
```

```python
stock_df.dropna(axis=0, inplace=True)
```

```python
price_df = pd.DataFrame(stock_df[["Open"]])
price_df.columns = ["Price"]
```

```python
plt.figure(figsize=(15,3))
plt.plot(price_df)
plt.title("Infy price", fontsize=15)
```

### there is no data for weekends and holidays, so a weekly average makes more sense 

```python
price_df=price_df.resample("W").mean()
price_df
```

```python
plt.plot(price_df)
seasonal_decompose(price_df).plot()
plt.show()
```

### plotting the acf and pacf, from the acf plot it can be seen that the current data is not stationary

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,3))
plot_acf(price_df, ax=ax[0])
plot_pacf(price_df, ax=ax[1])
plt.show()
```

```python
d = -1
is_stationary = False
diff_series = np.array([])
while not is_stationary:
    d += 1
    diff_series = np.diff(price_df["Price"], d)
    adf_test = adfuller(diff_series)
    if adf_test[1] < 0.05:
        is_stationary = True
```

### from pacf plot we can assume p=2, from acf plot we can assume q=1

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(18,3))
plot_acf(diff_series, ax=ax[0])
plot_pacf(diff_series, ax=ax[1])
plt.show()
```

```python
test_size_week = 30
train_size_week = price_df.shape[0]-test_size_week
```

```python
sarima_model = SARIMAX(price_df[:train_size_week], order=(2,d,1), seasonal_order=(2,d,1,7), enforce_stationarity=False, 
                enforce_invertibility=False)
sarima_model_fit = sarima_model.fit(disp=False)
```

```python
predictions = sarima_model_fit.predict(start=train_size_week, end=price_df.shape[0] + 30)
predictions.index = pd.date_range(start=price_df.index[train_size_week], periods=len(predictions), freq="W")   #price_df.index[train_size_week-1:]
```

```python
plt.figure(figsize=(15,5))
plt.plot(price_df[:train_size_week], label="train data")
plt.plot(price_df[train_size_week:], label="test data")
plt.plot(predictions, label="Prediction")
plt.legend()
```

```python
rms = np.sqrt(np.mean((predictions-price_df.iloc[train_size_week:]["Price"])**2))
print("root mean square value of predictions = ", rms)
```

```python
from pmdarima.arima import auto_arima
```

```python
stepwise_model = auto_arima(price_df[:train_size_week], start_p=2, start_q=1, d=1, D=1, start_P=2, start_Q=1, m=49,
                           trace=True, stepwise=True)
```

```python
future_forecast = pd.DataFrame(stepwise_model.predict(n_periods=test_size_week + 30))
future_forecast.index = pd.date_range(start=price_df.index[train_size_week], periods=len(future_forecast), freq="W")
future_forecast.columns = ["Price"]
```

```python
plt.figure(figsize=(15,5))
plt.plot(price_df[:train_size_week], label="train data")
plt.plot(price_df[train_size_week:], label="test data")
plt.plot(future_forecast, label="Prediction")
plt.legend()
```

```python
rms = np.sqrt(np.mean((future_forecast["Price"]-price_df.iloc[train_size_week:]["Price"])**2))
print("root mean square value of predictions = ", rms)
```

### updating the trained model with the test data

```python
stepwise_model.update(price_df[train_size_week:price_df.shape[0]-1])
```

```python
updated_future_forecast = pd.DataFrame(stepwise_model.predict(n_periods=30))
updated_future_forecast.index = pd.date_range(start=price_df.index[price_df.shape[0]-1], periods=len(updated_future_forecast), freq="W")
updated_future_forecast.columns = ["Price"]

plt.figure(figsize=(15,5))
plt.plot(price_df[:train_size_week], label="train data")
plt.plot(price_df[train_size_week:], label="test data")
plt.plot(updated_future_forecast, label="Prediction")
plt.legend()
```

### dump the model using pickle

```python
import pickle
model_filename = 'stock_prediction_model.pkl'
pickle.dump(stepwise_model, open(model_filename,'wb'))

pickle.dump(price_df.index[price_df.shape[0]-1], open('prev_trained_week.pkl','wb'))
```
